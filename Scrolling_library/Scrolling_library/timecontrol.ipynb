```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the factors and their categories
factors = {
    "Environmental Factors": [
        "Light and Darkness",
        "Weather and Seasons",
        "Urban Density and Architecture"
    ],
    "Social and Cultural Factors": [
        "Social Interactions",
        "Cultural Norms and Practices",
        "Public Events and Activities"
    ],
    "Psychological Factors": [
        "Stress and Anxiety",
        "Attention and Focus",
        "Routine and Novelty"
    ],
    "Technological Factors": [
        "Digital Devices and Media Consumption",
        "Transportation and Mobility",
        "Work and Productivity Tools"
    ],
    "Economic Factors": [
        "Work Schedules and Employment",
        "Economic Activity and Pace of Life"
    ],
    "Health and Well-being Factors": [
        "Physical Health",
        "Mental Health"
    ],
    "Sensory Factors": [
        "Noise Levels",
        "Smells and Scents"
    ]
}

# Create a directed graph
G = nx.DiGraph()

# Add nodes and edges
for category, subcategories in factors.items():
    G.add_node(category)
    for subcategory in subcategories:
        G.add_node(subcategory)
        G.add_edge(category, subcategory)

# Draw the graph
pos = nx.spring_layout(G, k=1.5, iterations=50)
plt.figure(figsize=(12, 12))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold', arrows=False)
plt.title("Factors Affecting Time Perception in Cities")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of individuals
num_individuals = 100

# Simulate time estimates for different time units
# For simplicity, we'll assume the true duration is always 1 unit

# Estimates for 1 minute
minute_estimates = np.random.normal(loc=1, scale=0.1, size=num_individuals)

# Estimates for 1 hour
hour_estimates = np.random.normal(loc=1, scale=0.3, size=num_individuals)

# Estimates for 1 day
day_estimates = np.random.normal(loc=1, scale=0.5, size=num_individuals)

# Calculate variances
minute_variance = np.var(minute_estimates)
hour_variance = np.var(hour_estimates)
day_variance = np.var(day_estimates)

# Plot the results
time_units = ['Minute', 'Hour', 'Day']
variances = [minute_variance, hour_variance, day_variance]

plt.figure(figsize=(10, 6))
plt.bar(time_units, variances, color=['blue', 'orange', 'green'])
plt.xlabel('Time Unit')
plt.ylabel('Variance in Estimates')
plt.title('Variance in Time Perception as Time Unit Increases')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Number of individuals
num_individuals = 100

# Simulate time estimates for different time units
# For simplicity, we'll assume the true duration is always 1 unit

# Estimates for 1 minute
minute_estimates = np.random.normal(loc=1, scale=0.1, size=num_individuals)

# Estimates for 1 hour
hour_estimates = np.random.normal(loc=1, scale=0.3, size=num_individuals)

# Estimates for 1 day
day_estimates = np.random.normal(loc=1, scale=0.5, size=num_individuals)

# Calculate variances
minute_variance = np.var(minute_estimates)
hour_variance = np.var(hour_estimates)
day_variance = np.var(day_estimates)

# Combine all data for easier plotting
data = {
    'Minute': minute_estimates,
    'Hour': hour_estimates,
    'Day': day_estimates
}

# Convert data to a suitable format for seaborn
import pandas as pd
df = pd.DataFrame(data)

# Plot the results
plt.figure(figsize=(15, 5))

# Bar Plot
plt.subplot(1, 3, 1)
time_units = ['Minute', 'Hour', 'Day']
variances = [minute_variance, hour_variance, day_variance]
plt.bar(time_units, variances, color=['blue', 'orange', 'green'])
plt.xlabel('Time Unit')
plt.ylabel('Variance in Estimates')
plt.title('Variance in Time Perception')

# Box Plot
plt.subplot(1, 3, 2)
sns.boxplot(data=df)
plt.xlabel('Time Unit')
plt.ylabel('Estimated Time')
plt.title('Box Plot of Time Estimates')

# Violin Plot
plt.subplot(1, 3, 3)
sns.violinplot(data=df)
plt.xlabel('Time Unit')
plt.ylabel('Estimated Time')
plt.title('Violin Plot of Time Estimates')

plt.tight_layout()
plt.show()
```
