```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time range for Monday to Friday
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
hours_per_day = 24
total_hours = len(days_of_week) * hours_per_day

# Create an array of time points across the five days
time = np.linspace(0, total_hours, 1000)

# Define the frequency (one full cycle per day)
frequency = 2 * np.pi / hours_per_day

# Amplitude (representing the height of the sun above the horizon)
amplitude = 1  # Normalized to 1 for simplicity

# Calculate the sine wave representing the sun cycle
sun_cycle = amplitude * np.sin(frequency * time)

# Class schedule (start and end times in hours from the start of the week, with locations)
classes = {
    'Monday': [(12, 16, 'Buell'), (18.5, 20.5, 'Buell')],   # 12:00-4:00, 6:30-8:30
    'Tuesday': [(9, 11, 'Avery'), (18, 20, 'Buell')],       # 9:00-11:00, 6:00-8:00
    'Wednesday': [],                                        # No classes
    'Thursday': [(12, 16, 'Buell'), (18, 20, 'Buell')],     # 12:00-4:00, 6:00-8:00
    'Friday': [(9, 11, 'Avery')]                            # 9:00-11:00
}

# Travel times (assuming 30 minutes to arrive at and leave each location)
travel_times = {
    'Buell': 0.5,  # 30 minutes to arrive/leave
    'Avery': 0.5   # 30 minutes to arrive/leave
}

# Define font properties (Roboto Mono)
font_title = {'family': 'Roboto Mono', 'weight': 'bold', 'size': 18}
font_labels = {'family': 'Roboto Mono', 'size': 14}
font_legend = {'family': 'Roboto Mono', 'size': 12}

# Define custom colors
sun_color = '#FFA500'  # Orange for the sun
night_color = '#1C1C1C'  # Darker color for night
day_color = '#87CEEB'  # Sky blue for day
avery_color = '#0000FF'  # Blue for Avery
buell_color = '#FFFF00'  # Yellow for Buell
apartment_color = '#FF0000'  # Red for Apartment
marker_size = 100  # Marker size for arrival and leaving times

# Plot the sun cycle
plt.figure(figsize=(14, 7))
plt.plot(time, sun_cycle, label="Sun Cycle (Sine Wave)", color=sun_color, linewidth=2)

# Shade the night areas (where sun_cycle is negative)
plt.fill_between(time, sun_cycle, where=(sun_cycle < 0), color=night_color, alpha=0.3, label='Night')

# Shade the day areas (where sun_cycle is positive)
plt.fill_between(time, sun_cycle, where=(sun_cycle >= 0), color=day_color, alpha=0.2, label='Day')

# Shade apartment time from 11:00 PM to 6:00 AM
for day_index in range(len(days_of_week)):
    start_time = day_index * hours_per_day + 23  # 11:00 PM
    end_time = (day_index + 1) * hours_per_day + 6  # 6:00 AM
    plt.fill_betweenx([-1, 1], start_time, end_time, color=apartment_color, alpha=0.2, label='Time at Apartment' if day_index == 0 else "")

# Plot the class schedule and locations on the same diagram
location_colors = {'Avery': avery_color, 'Buell': buell_color, 'Apartment': apartment_color}

for day, periods in classes.items():
    day_index = days_of_week.index(day)
    for start, end, location in periods:
        start_time = day_index * hours_per_day + start
        end_time = day_index * hours_per_day + end
        
        # Plot time spent in class
        plt.fill_betweenx([-1, 1], start_time, end_time, color=location_colors[location], alpha=0.5, label=f'{location} Class')
        
        # Plot arrival time
        arrival_time = start_time - travel_times[location]
        plt.scatter(arrival_time, 0, color=location_colors[location], s=marker_size, label=f'Arrive at {location}' if day == 'Monday' and start == 12 else "")
        
        # Plot leaving time
        leaving_time = end_time + travel_times[location]
        plt.scatter(leaving_time, 0, color=location_colors[location], s=marker_size, label=f'Leave {location}' if day == 'Monday' and start == 12 else "")

# Customize x-axis with days of the week (Monday to Friday)
plt.xticks(ticks=np.arange(0, total_hours, hours_per_day), labels=days_of_week, **font_labels)

# Set y-axis labels to show day/night context
plt.yticks([-1, 0, 1], ['Night', 'Horizon', 'Day'], **font_labels)

# Label the axes
plt.xlabel("Day of the Week", **font_labels)
plt.ylabel("Sun Position", **font_labels)

# Add a title
plt.title("Class Schedule with Locations, Apartment Time, and Travel Times Overlaid on Sun Cycle (Monday to Friday)", **font_title)

# Add grid and legend (avoid duplicate labels)
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), loc='upper right', prop=font_legend)

# Save the plot as an SVG
plt.savefig("class_schedule_sun_cycle_travel_apartment.svg", format="svg")

# Show the plot
plt.show()
```
