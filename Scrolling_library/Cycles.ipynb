```python
import numpy as np
import matplotlib.pyplot as plt

# Define the timeline (in years) for 20 years
years = np.linspace(0, 20, 1000)

# Define the cycles with their corresponding periods (in years)
cycles = {
    'Grasshopper Abundance': 9.2,
    'Sunspots and Rainfall': 12,
    'Bean Seed Water Uptake': 7.4 / 365,  # converting days to years
    'UFO Sightings': 61 / 12,  # converting months to years
    'Bank Deposits and Milk Cows': 9,
    'Canadian Animal Abundance': 9.6,
    'International War Battles': 6,
    'Steel Production and Sugar Prices': 6,
    'US Precipitation and Cigarette Production': 8,
    'Nile River Floods': 17.333  # 17 1/3 years
}

# Create a figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Plot each cycle
for event, period in cycles.items():
    # Create a sinusoidal wave for each cycle
    cycle_wave = np.sin(2 * np.pi * years / period)
    ax.plot(years, cycle_wave, label=f'{event} ({period:.2f} years)')

# Add labels and legend
ax.set_title('Cyclic Patterns Over Time')
ax.set_xlabel('Years')
ax.set_ylabel('Normalized Amplitude')
ax.legend(loc='upper right', bbox_to_anchor=(1.4, 1))

# Limit the y-axis to keep the plot from stretching too much
ax.set_ylim(-2, 2)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the timeline (in years) for 25 years
years = np.linspace(0, 25, 1000)

# Define the cycles with their corresponding periods (in years)
cycles = {
    'Grasshopper Abundance': 9.2,
    'Sunspots and Rainfall': 12,
    'Bean Seed Water Uptake': 7.4 / 365,  # converting days to years
    'UFO Sightings': 61 / 12,  # converting months to years
    'Bank Deposits and Milk Cows': 9,
    'Canadian Animal Abundance': 9.6,
    'International War Battles': 6,
    'Steel Production and Sugar Prices': 6,
    'US Precipitation and Cigarette Production': 8,
    'Nile River Floods': 17.333  # 17 1/3 years
}

# Create a figure and axis
fig, ax = plt.subplots(len(cycles), 1, figsize=(14, 20), sharex=True)

# Plot each cycle individually
for i, (event, period) in enumerate(cycles.items()):
    # Create a sinusoidal wave for each cycle
    cycle_wave = np.sin(2 * np.pi * years / period)
    ax[i].plot(years, cycle_wave, label=f'{event} ({period:.2f} years)')
    ax[i].set_title(f'{event} ({period:.2f} years)')
    ax[i].set_ylim(-1.5, 1.5)
    ax[i].grid(True)
    ax[i].legend()

# Add common labels
fig.suptitle('Cyclic Patterns Over 25 Years', fontsize=16)
plt.xlabel('Years')
plt.ylabel('Normalized Amplitude')
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Display the plot
plt.show()
```
